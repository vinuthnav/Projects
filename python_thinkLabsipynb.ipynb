```python
def extract_real(value):
    try:
        # If the value is already a complex number
        if isinstance(value, complex):
            return value.real

        # Handle strings with complex numbers
        elif isinstance(value, str):
            value = value.strip("()")  # Remove enclosing parentheses
            if "j" in value:  # Indicates a complex number
                real_part = value.split("+")[0] if "+" in value else value.split("-")[0]
                return float(real_part.strip())

        # If the value is a regular number
        return float(value)
    except Exception as e:
        print(f"Error parsing value {value}: {e}")
        return None
```
